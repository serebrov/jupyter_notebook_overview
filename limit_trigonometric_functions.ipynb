# $\tan(3 - \sqrt{9 - x^3})$

```python
expression = tan(3 - sqrt(9 - x*x*x))
expression
```

```python
plot(expression, (x, -2.25, 1.9), show=True)
```

```python
plot(expression, (x,-0.01, 0.01), show=True)
```

```python
limit(expression, x, 0)
```

# $\bar\exists \lim_{x\to0} \sin(\frac{1}{x})$

The limit for $sin(\frac{1}{x})$ when $x\to0$ does not exist

 $$\bar\exists \lim_{x\to0} \sin(\frac{1}{x})$$

```python
plot((sin(1/x)), (x,-1,1))
```

The $\frac{1}{x}$ grows by module to $\inf$ when $x\to0$, but $sin(x)$ just continues oscillating when $x\to\inf$.

```python
limit(1/x, x, 0)
```

```python
plot((1/x),(x,-1,1), ylim=(-100,100))
```

# $\lim_{x\to\pi} \frac{\sin{3x} (\cos^2{5x})}{e^x - e^\pi}$

```python
expression = ((sin(3*x)*(cos(5*x)**2)))/(E**x - E**pi)
expression
```

```python
limit(expression, x, pi)
```

# $\lim_{x\to0} \frac{e^x-1}{\ln(1+x)}$

```python
expression = (E**x-1)/(log(1+x, E))
expression
```

```python
limit(expression, x, 0)
```
